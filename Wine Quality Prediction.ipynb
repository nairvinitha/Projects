```python
#To decide if the wine quality is good or not, we should put it in two categories. 'Good wine' and 'Bad wine'. Let these classes
#be assigned as 0 and 1. Therefore we should binarize our target variable.

#for x in df['quality']:
 #   if x >= 7:
  #      df['quality'] = 1
   # else:
    #    df['quality'] = 0
        
    #return df['quality']
    
df['quality'] = [1 if x >= 7 else 0 for x in df.quality]
print(df['quality'].unique())
```

Splitting the data into training and testing sets:

```python
#Get all the feature variables into a separate feature matrix, vector x and the target/response variable into separate matrix 
#vector y

x = df.drop(['quality'], axis = 1)
y = df['quality']
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size = 0.2, random_state = 40)
```

```python
#to avoid model bias, we need to apply feature scaling so that features with high values do not dominate the distance 
#values between data points
from sklearn.preprocessing import StandardScaler
StdScale = StandardScaler()
train_x = StdScale.fit_transform(train_x)
test_x = StdScale.fit_transform(test_x)
train_x
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn. tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error
import numpy as np
#creating RandomForestClassifier constructor
RFC = RandomForestClassifier()
# fit data
RFC.fit(train_x, train_y)
# predicting score
RFC_score = RFC.score(test_x,test_y)
print('Model performance score : ',RFC_score)
#predicting on the test set
y_pred1 = RFC.predict(test_x)
#get the classification report
print(classification_report(test_y, y_pred1))
```

```python
#checking for Support Vector Machine
SVM = SVC(kernel='rbf')
# fit data
SVM.fit(train_x, train_y)
# predicting score
SVM_score = SVM.score(test_x,test_y)
print('Model performance score : ',SVM_score)
#predicting on the test set
y_pred2 = SVM.predict(test_x)
#get the classification report
print(classification_report(test_y, y_pred2))
```

```python
#checking for logistic regression model
LR = LogisticRegression()
# fit data
LR.fit(train_x, train_y)
# predicting score
LR_score = LR.score(test_x,test_y)
print('Model performance score : ',LR_score)
#predicting on the test set
y_pred3 = LR.predict(test_x)
#get the classification report
print(classification_report(test_y, y_pred3))
```

```python
#checking for Decision Tree Classifier 
DTC =  DecisionTreeClassifier()
# fit data
DTC.fit(train_x, train_y)
# predicting score
DTC_score = DTC.score(test_x,test_y)
print('Model performance score : ',DTC_score)
#predicting on the test set
y_pred4 = DTC.predict(test_x)
#get the classification report
print(classification_report(test_y, y_pred4))
```
